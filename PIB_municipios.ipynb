```python
import pandas as pd
import matplotlib.pyplot as plt

# === 1. Leitura e limpeza dos dados ===
path = "/Users/junior/Downloads/tabelas_completas.xlsx"
df = pd.read_excel(path, sheet_name="Tabela 4", header=4)

# Remover primeira linha de notas e renomear colunas
df = df.drop(0)
df = df.rename(columns={
    "Popula√ß√£o\n (2)": "Popula√ß√£o",
    "Produto\nInterno Bruto\nper capita\n(1) (R$)": "PIB_per_capita"
})

# Converter colunas para num√©rico
df["Popula√ß√£o"] = pd.to_numeric(df["Popula√ß√£o"], errors="coerce")
df["PIB_per_capita"] = pd.to_numeric(df["PIB_per_capita"], errors="coerce")

# Remover linhas com valores ausentes
df = df.dropna(subset=["PIB_per_capita", "Popula√ß√£o"]).reset_index(drop=True)

print("Pr√©-processamento conclu√≠do ‚úÖ")
print(df.head())

```

```python
# === 2. Estat√≠sticas descritivas ===
print("\nüìä Estat√≠sticas descritivas do PIB per capita:")
print(df["PIB_per_capita"].describe())

# Histograma
plt.figure(figsize=(8,5))
df["PIB_per_capita"].hist(bins=30, edgecolor="black")
plt.title("Distribui√ß√£o do PIB per capita")
plt.xlabel("PIB per capita (R$)")
plt.ylabel("Frequ√™ncia")
plt.show()

# Boxplot
plt.figure(figsize=(6,4))
df["PIB_per_capita"].plot(kind="box")
plt.title("Boxplot do PIB per capita")
plt.show()
```
